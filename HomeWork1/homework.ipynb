```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
df = pd.read_csv ('hw1_std.csv')
print (df)
```

```python
df['sex'] = df['sex'].replace({'F': '0'}) 
df['sex'] = df['sex'].replace({'M': '1'}) 

print(df)
```

```python
df.fillna(value=0, inplace=True)

print(df)
```

```python
df[df.duplicated(keep=False)==True]
df.drop_duplicates(keep='first')
```

```python
df.mean()
```

```python
df.std()
```

```python
df.count()
```

```python
df.corr()
```

```python
ax = sns.scatterplot(x="ord1",y="target", data=df)
ax.set_title('target & ord1')

```

```python
ax = sns.scatterplot(x="num3",y="target", data=df)
ax.set_title('target & num3')
```

In sex a graph shows that female have more percentage to get disease. But for age can see that male have high percentage to get diseases when they are young but female is opposite.

```python
sns.catplot(x='age', y='sex', hue='target', kind='bar', palette="pastel", data=df)
plt.title('AGE vs SEX')
```

I can tell that if ord1 is 0 we can assume that they not get a diseases but if ord1 is 2 we can assume that they are get dissease.

```python
g = sns.jointplot(x='ord1', y='target',data = df, kind='kde', color='blue')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels('$ord1$', '$target$')
ax.set_title('target & ord1')
```

I can tell that if num3 is 140 we can assume that they not get a diseases but if num3 is 165 we can assume that they are get dissease.

```python
g = sns.jointplot(x='num3', y='target',data = df, kind='kde', color='green')
g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels('$num3$', '$target$')
ax.set_title('target & num3')
```
